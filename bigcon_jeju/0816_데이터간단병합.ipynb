```python
import pandas as pd
import numpy as np
```

### 음쓰 전처리

```python
df1 = pd.read_csv('01_음식물쓰레기_FOOD_WASTE.CSV.csv', encoding='cp949')
df1.head()
```

```python
# 알수없음/우도면/추자면 제거
#df1 = df1[df1['emd_nm'] != '알수없음']
df1 = df1[df1['emd_nm'] != '우도면']
df1 = df1[df1['emd_nm'] != '추자면']
len(df1)
```

```python
# base_date 날짜형 변환
df1['base_date'] = pd.to_datetime(df1['base_date'])

# Timestamp를 Period로 변환 - 월별 
dates = list(df1['base_date'])
ts_dates = pd.to_datetime(dates)
pr_month = ts_dates.to_period(freq='M')

# 시계열 데이터 - 월별 인덱스 
음쓰 = df1.set_index(pr_month)
음쓰.head() 
```

```python
#음쓰.to_csv('음쓰_월별.csv', encoding='cp949')
```

```python
# 일별 + 음면동 별 
음쓰_일별 = 음쓰.groupby(['base_date', 'emd_nm']).sum()
음쓰_일별
```

```python
# 월별 + 읍면동 별
음쓰['month'] = pr_month 
음쓰_월별 = 음쓰.groupby(['month', 'emd_nm']).sum()
음쓰_월별
```

### 거주인구 전처리

```python
df5 = pd.read_csv('03_거주인구_RESIDENT_POP.CSV.csv', encoding='cp949')
df5.tail()
```

```python
# 알수없음/우도면/추자면 제거
df5 = df5[df5['emd_nm'] != '알수없음']
df5 = df5[df5['emd_nm'] != '우도면']
df5 = df5[df5['emd_nm'] != '추자면']
len(df5)
```

```python
# base_date = base_year + base_month
df5['base_year'] = df5['base_year'].astype('str')
df5['base_month'] = df5['base_month'].astype('str')

df5['base_year'] = df5['base_year'].astype('str')
df5['base_month'] = df5['base_month'].astype('str')
df5['base_date'] = df5[['base_year', 'base_month']].apply("-".join, axis=1)
```

```python
# base_date: 날짜형으로 변환 -> 이미 월별로 묶임 
df5['base_date'] = pd.to_datetime(df5['base_date'])

# 시계열 데이터 변환
거주인구 = df5.set_index('base_date')

# Timestamp를 Period로 변환 - 월별 
dates = list(df5['base_date'])
ts_dates = pd.to_datetime(dates)
pr_month = ts_dates.to_period(freq='M')
거주인구 = df5.set_index(pr_month)
거주인구.tail()
```

```python
#거주인구.to_csv('거주인구_월별.csv', encoding='cp949')
```

```python
# 월별 + 읍면동별
거주인구['month'] = 거주인구['base_date'].dt.to_period(freq='M')
거주인구.drop(['base_year','base_month','base_date','emd_cd'], axis=1, inplace=True)
거주인구_월별 = 거주인구.groupby(['month', 'emd_nm']).sum()
거주인구_월별
```

### 카드소비 전처리

```python
df6 = pd.read_csv('04_음식관련 카드소비_CARD_SPENDING.CSV.csv', encoding='cp949')
df6.tail()
```

```python
# 카드 데이터 최근이 없는 줄 알았는데 걍 정렬이 안된 거였음(머쓱^^;;)
df6['base_date'] = pd.to_datetime(df6['base_date'])
df6 = df6.sort_values(by='base_date')
```

```python
# 알수없음/우도면/추자면 제거
df6 = df6[df6['emd_nm'] != '알수없음']
df6 = df6[df6['emd_nm'] != '우도면']
df6 = df6[df6['emd_nm'] != '추자면']
len(df6)
```

```python
# base_date 날짜형 변환
df6['base_date'] = pd.to_datetime(df6['base_date'])

# 카드 카테고리 수정 for 가정용
분류 = ['농축수산물', '마트/슈퍼마켓', '식품', '간식', '배달']
df6 = df6.query("mct_cat_nm in @분류")

# Timestamp를 Period로 변환 - 월별 
dates = list(df6['base_date'])
ts_dates = pd.to_datetime(dates)
pr_month = ts_dates.to_period(freq='M')

# 시계열 데이터 - 월별 인덱스 
카드 = df6.set_index(pr_month)
카드.head() 
```

```python
#카드.to_csv('카드_월별.csv', encoding='cp949')
```

```python
# 일별 + 음면동 별 
카드_일별 = 카드.groupby(['base_date', 'emd_nm']).sum()
카드_일별
```

```python
# 월별 + 읍면동 별
카드['month'] = pr_month 
카드_월별 = 카드.groupby(['month', 'emd_nm']).sum()
카드_월별
```

### 데이터 병합
* 이건 어차피 다시 해야겠지만 일단 있는 걸로 병합함

```python
# left
df_left = pd.merge(음쓰_월별, 거주인구_월별, how='left', on=['month', 'emd_nm'])
df_left = pd.merge(df_left, 카드_월별, how='left', on=['month', 'emd_nm'])
df_left
```

```python
df_left.to_csv('음쓰거주인구카드병합.csv', encoding='cp949')
```

```python
# outer: 음쓰 배출량이 결측으로 뜨니까 이걸 있는 변수로 예측해봐도?
df_outer = pd.merge(음쓰_월별, 거주인구_월별, how='outer', on=['month', 'emd_nm'])
df_outer = pd.merge(df_outer, 카드_월별, how='outer', on=['month', 'emd_nm'])
df_outer
```
